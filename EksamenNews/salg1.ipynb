```python
import pandas as pd
import matplotlib.pyplot as plt

# Organiser dataene som en liste av lister
data = [
    ["2019", 3681323, 1765675, 68060, 0],
    ["2020", 6055351, 20894519, 4121351, 0],
    ["2021", 14386061, 50506996, 88383206, 0],
    ["2022", 37678938, 115914923, 88383206, 0],
    ["2023", 51532875, 259204605, 179537544, 44362077]

]
# Lag en DataFrame med Pandas
df = pd.DataFrame(data, columns=["År", "Mysimba", "Ozempic", "Saxenda", "Wegovy"])


# Konverter kolonnene til numeriske verdier
df[["Mysimba", "Ozempic", "Saxenda", "Wegovy"]] = df[["Mysimba", "Ozempic", "Saxenda", "Wegovy"]].apply(pd.to_numeric)

# Plot grafen
#plt.figure(figsize=(10, 6))
#plt.plot(df["År"], df["Mysimba"], label="Mysimba", marker='o')
#plt.plot(df["År"], df["Ozempic"], label="Ozempic", marker='o')
#plt.plot(df["År"], df["Saxenda"], label="Saxenda", marker='o')
#plt.plot(df["År"], df["Wegovy"], label="Wegovy", marker='o')
plt.figure(figsize=(10, 6))
plt.bar(df["År"], df["Mysimba"], label="Mysimba")
plt.bar(df["År"], df["Ozempic"], label="Ozempic", bottom=df["Mysimba"])
plt.bar(df["År"], df["Saxenda"], label="Saxenda", bottom=df["Mysimba"] + df["Ozempic"])
plt.bar(df["År"], df["Wegovy"], label="Wegovy", bottom=df["Mysimba"] + df["Ozempic"] + df["Saxenda"])


# Legg til tittel og aksetitler
plt.title('Samlet omsetning av slankepreparater')
plt.xlabel('År')
plt.ylabel('100 millioner kroner')

# Vis legenden
plt.legend()

# Sett y-aksen til å starte fra 0
plt.savefig("salg_produkter.svg", format="svg")


# Vis grafen
plt.show()
```
